```python
# a. Count the number of unique tokens in the text.

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

paragraph = "On the 27th of March, 2000, John Smith was born. His email address is john.smith@email.com, and his phone number is 555-1234. John is a friendly person, but he doesn't enjoy crowded places. He prefers quiet environments. However, don't mistake his reserved nature for unfriendliness. John is an excellent communicator, and his punctuality is noteworthy. His date of birth, 27-March-2000, is a special day for him, and he often celebrates it with close friends and family."

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word.lower() for word in tokens if word.isalnum() and word.lower() not in stop_words]

# Count the number of unique tokens
unique_tokens = set(filtered_tokens)
num_unique_tokens = len(unique_tokens)

print("Number of unique tokens:", num_unique_tokens)
```

```python
# b. Count the number of punctuation characters and remove punctuations.

import string

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Remove punctuation and convert to lowercase
table = str.maketrans('', '', string.punctuation)
stripped = [word.translate(table).lower() for word in tokens if word.isalnum()]

# Count the number of punctuation characters
num_punctuation = sum(1 for c in paragraph if c in string.punctuation)

print("Number of punctuation characters:", num_punctuation)
print("Paragraph without punctuations:", ' '.join(stripped))
```

```python
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

# Sample paragraph for POS tagging
paragraph = "Your sample paragraph goes here."

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words]

# Perform POS tagging
pos_tags = pos_tag(filtered_tokens)

# Extract the POS tags
tags = [tag for word, tag in pos_tags]

# Create a frequency distribution of POS tags
tag_freq = nltk.FreqDist(tags)

# Plot the distribution of each POS tag using a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x=list(tag_freq.keys()), palette="viridis")
plt.title('Distribution of POS Tags')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.show()

# Print a sample of tokens and their corresponding POS tags
for index in range(min(10, len(pos_tags))):
    print(pos_tags[index])

```

```python
# f. Determine the number of unique lemma available in the text.

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

# Given paragraph
paragraph = "On the 27th of March, 2000, John Smith was born. His email address is john.smith@email.com, and his phone number is 555-1234. John is a friendly person, but he doesn't enjoy crowded places. He prefers quiet environments. However, don't mistake his reserved nature for unfriendliness. John is an excellent communicator, and his punctuality is noteworthy. His date of birth, 27-March-2000, is a special day for him, and he often celebrates it with close friends and family."

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the tokens and find unique lemmas
lemmas = [lemmatizer.lemmatize(token) for token in tokens]
unique_lemmas = set(lemmas)

# Number of unique lemmas
num_unique_lemmas = len(unique_lemmas)
print("Number of unique lemmas:", num_unique_lemmas)
```

```python
# g. Plot the frequency distribution of words in the text. Display only the top 10
# (most occurring) tokens in the chart.

from nltk import FreqDist
from nltk.tokenize import word_tokenize

# Given paragraph
paragraph = "On the 27th of March, 2000, John Smith was born. His email address is john.smith@email.com, and his phone number is 555-1234. John is a friendly person, but he doesn't enjoy crowded places. He prefers quiet environments. However, don't mistake his reserved nature for unfriendliness. John is an excellent communicator, and his punctuality is noteworthy. His date of birth, 27-March-2000, is a special day for him, and he often celebrates it with close friends and family."

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Calculate the frequency distribution of words
fdist = FreqDist(tokens)

# Select the top 10 most occurring tokens
top_10_tokens = fdist.most_common(10)

# Plot the frequency distribution of words for the top 10 tokens
plt.figure(figsize=(10, 6))
fdist.plot(10, title='Top 10 Most Common Words')
plt.show()

```

```python
# h. Find the number of unique bigrams, trigrams and quadgrams (n=4) in the corpus
from nltk import bigrams, trigrams, ngrams

# Given paragraph
paragraph = "On the 27th of March, 2000, John Smith was born. His email address is john.smith@email.com, and his phone number is 555-1234. John is a friendly person, but he doesn't enjoy crowded places. He prefers quiet environments. However, don't mistake his reserved nature for unfriendliness. John is an excellent communicator, and his punctuality is noteworthy. His date of birth, 27-March-2000, is a special day for him, and he often celebrates it with close friends and family."

# Tokenize the paragraph
tokens = word_tokenize(paragraph)

# Generate bigrams, trigrams, and quadgrams
bi_grams = list(bigrams(tokens))
tri_grams = list(trigrams(tokens))
quad_grams = list(ngrams(tokens, 4))

# Count the number of unique bigrams, trigrams, and quadgrams
num_unique_bigrams = len(set(bi_grams))
num_unique_trigrams = len(set(tri_grams))
num_unique_quadgrams = len(set(quad_grams))

print("Number of unique bigrams:", num_unique_bigrams)
print("Number of unique trigrams:", num_unique_trigrams)
print("Number of unique quadgrams:", num_unique_quadgrams)
```

```python
# i. Find all dates and convert them to the DD-MM-YYYY format.
import re

# Define a regular expression pattern to match dates in the format: 27-March-2000
date_pattern = r'\b(\d{1,2})-(\w+)-(\d{4})\b'

# Function to convert the matched dates to DD-MM-YYYY format
def convert_to_dd_mm_yyyy(match):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    day, month, year = match.groups()
    month_number = months.index(month) + 1
    return f"{day.zfill(2)}-{str(month_number).zfill(2)}-{year}"

# Find all dates and convert them to DD-MM-YYYY format
converted_paragraph = re.sub(date_pattern, convert_to_dd_mm_yyyy, paragraph)

print(converted_paragraph)
```

```python
# j. Plot a distribution of the different values of year occurring in the text.


#  Define a regular expression pattern to match years in the format: YYYY
year_pattern = r'(19\d{2}|20\d{2})'  # Matches years from 1900 to 2099

# Find all years in the paragraph
years = re.findall(year_pattern, paragraph)

# Create a frequency distribution of years
year_distribution = {}
for year in years:
    if year in year_distribution:
        year_distribution[year] += 1
    else:
        year_distribution[year] = 1

# Plot the distribution of different years
plt.figure(figsize=(10, 6))
plt.bar(year_distribution.keys(), year_distribution.values(), color='skyblue')
plt.title('Distribution of Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```
